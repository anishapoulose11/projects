**Step 2: Load the dataset**

```python
full_df = pd.read_csv('/content/1429_1.csv', nrows=3000)
df = full_df[["reviews.text","reviews.title"]]
df[["reviews.text","reviews.title"]] = df[["reviews.text","reviews.title"]].astype(str)
full_df.head()
```

**Tokenization:**

```python
import re

def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens

df['reviews.title_tokenized'] = df['reviews.title'].apply(lambda x: tokenize(x.lower()))

df['reviews.text_tokenized'] = df['reviews.text'].apply(lambda x: tokenize(x.lower()))

df.head()
```

**Lower Casing:**

```python
# drop the new column created in last cell
df.drop(["reviews.text_tokenized"], axis=1, inplace=True)
df.drop(["reviews.title_tokenized"], axis=1, inplace=True)
df["reviews.text_lower"] = df["reviews.text"].str.lower()
df["reviews.title_lower"] = df["reviews.title"].str.lower()
df.head()
```

**Removal of Punctuations:**

```python
#drop the new column created in last cell
df.drop(["reviews.text_lower"], axis=1, inplace=True)
df.drop(["reviews.title_lower"], axis=1, inplace=True)

PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    """custom function to remove the punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

df["reviews.text_wo_punct"] = df["reviews.text"].apply(lambda text: remove_punctuation(text))
df["reviews.title_wo_punct"] = df["reviews.title"].apply(lambda text: remove_punctuation(text))
df.head()
```

**Removal of stopwords:**

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
", ".join(stopwords.words('english'))
```

```python
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

df["reviews.text_wo_stop"] = df["reviews.text_wo_punct"].apply(lambda text: remove_stopwords(text))
df["reviews.title_wo_stop"] = df["reviews.title_wo_punct"].apply(lambda text: remove_stopwords(text))
df.head()
```

**Stemming**

```python
from nltk.stem.porter import PorterStemmer
```

```python
#Drop the two columns
df.drop(["reviews.text_wo_stop", "reviews.text_wo_punct"], axis=1, inplace=True)
df.drop(["reviews.title_wo_stop", "reviews.title_wo_punct"], axis=1, inplace=True)

stemmer = PorterStemmer()
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])

df["reviews.text_stemmed"] = df["reviews.text"].apply(lambda text: stem_words(text))
df["reviews.title_stemmed"] = df["reviews.title"].apply(lambda text: stem_words(text))
df.head()
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])

df["review.text_lemmatized"] = df["reviews.text"].apply(lambda text: lemmatize_words(text))

df["review.title_lemmatized"] = df["reviews.title"].apply(lambda text: lemmatize_words(text))
df.head()
```
